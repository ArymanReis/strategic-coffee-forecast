```python
import yfinance as yf
import pandas as pd
```

```python
# Definindo os ativos (Tickers)
# KC=F -> Coffee C Futures (Arábica - Contrato Global)
# BRL=X -> Dólar vs Real
tickers = ['KC=F', 'BRL=X']

print(f"Baixando dados para: {tickers}...")
# Baixando dados dos últimos 5 anos
dados = yf.download(tickers, period="5y", interval="1d")

# Visualizando
print("\n--- Amostra dos Dados ---")
display(dados.head())
```

```python
# Salvando os dados na pasta 'raw' (Dados Brutos)
# Explicação do caminho: '../' significa "volte uma pasta" (sair de notebooks para ir para data)
dados.to_parquet('../data/raw/dados_brutos_commodities.parquet')

print("✅ Arquivo salvo com sucesso! Pode verificar na pasta data/raw.")
```
